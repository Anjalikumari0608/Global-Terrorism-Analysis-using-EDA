````python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('/content/drive/MyDrive/Global Terrorism Data.csv', encoding='latin')
df.head()
````

````python
df = df[['iyear','country_txt','region_txt','attacktype1_txt','nkill','nwound']]

df.columns = ['Year','Country','Region','AttackType','Killed','Wounded']

df.fillna(0, inplace=True)

df.head()
````

````python
sns.countplot(x='Year', data=df)
plt.xticks(rotation=90)
plt.title("Attacks per Year")
plt.show()
````

This graph shows the number of terrorist attacks per year. It can be observed that the number of attacks was low in earlier years but gradually increased over time. A significant rise is seen after 2000, indicating growing global terrorism activities in recent years.

````python
df['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Affected Countries")
plt.xlabel("Country")
plt.ylabel("Number of Attacks")
plt.show()
````

This graph represents the top 10 countries most affected by terrorist attacks. Iraq has the highest number of attacks, followed by Pakistan and Afghanistan. This indicates that these regions are more prone to terrorism activities.

````python
df['AttackType'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Attack Type Distribution")
plt.ylabel("")
plt.show()
````

This pie chart represents the distribution of different types of terrorist attacks. It is observed that Bombing/Explosion is the most common type of attack, contributing the highest percentage. Armed Assault and Assassination are also significant. This indicates that explosive and violent attack methods are widely used in terrorism activities.

````python
sns.countplot(y='Region', data=df)
plt.title("Region-wise Attacks")
plt.show()
````

This graph shows the distribution of terrorist attacks across different regions. It is observed that the Middle East & North Africa region has the highest number of attacks, followed by South Asia. Regions like Sub-Saharan Africa and South America also show significant activity, while North America and Central Asia have comparatively lower attacks. This indicates that terrorism is more concentrated in specific regions.

````python
df[['Killed','Wounded']].sum().plot(kind='bar')
plt.title("Total Casualties")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
````

This graph compares the total number of people killed and wounded in terrorist attacks. It is observed that the number of wounded people is higher than the number of fatalities. This indicates that terrorist attacks cause severe injuries and impact a large number of people beyond deaths.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python

missing_percent = (df.isnull().sum() / len(df)) * 100
df = df.drop(columns=missing_percent[missing_percent > 50].index)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
  df[col] = df[col].fillna(df[col].mode()[0])
df = df.drop_duplicates()
df.reset_index(drop=True, inplace=True)
df.shape
````

### What all manipulations have you done and insights you found?

In the data wrangling process, missing values were handled by removing columns with more than 50% null values. Remaining missing values in numerical columns were filled using median, and categorical columns were filled using mode.

Duplicate records were checked and no significant duplicates were found, ensuring data quality. After cleaning, the dataset was reduced to 58 columns from the original dataset, making it more efficient for analysis.

Insights: The dataset is large with over 181,000 records and contains significant missing values initially. After cleaning, the data became more structured and reliable, making it suitable for further analysis and visualization.
